```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Dados hipotéticos
dados = {'Palavras_Chave': [3, 7, 2, 8, 1, 5],
         'Spam': [0, 1, 0, 1, 0, 1]}
df = pd.DataFrame(dados)

# Treinamento do modelo
X = df[['Palavras_Chave']]
y = df['Spam']
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X, y)

# Gráfico
plt.scatter(df['Palavras_Chave'], df['Spam'], c=df['Spam'], cmap='viridis')
plt.xlabel('Número de Palavras-Chave')
plt.ylabel('Spam (1) ou Não Spam (0)')
plt.title('Classificação de E-mails')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA

# Dados hipotéticos
dados = {
    'Texto': [
        'Promoção imperdível de viagens',
        'Você ganhou um cartão de crédito grátis',
        'Curso de programação em Python',
        'Participação em sorteio de prêmios',
        'Convite para evento de tecnologia',
        'Ganhou bitcoins gratuitos agora'
    ],
    'Spam': [1, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(dados)

# Vetorização do texto
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Texto'])

# Conversão para array
X_array = X.toarray()
y = df['Spam']

# Treinamento do modelo
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X_array, y)

# Redução de dimensionalidade para visualização
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_array)

# Gráfico
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Classificação de E-mails Baseada em Palavras-Chave')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA

# Dados hipotéticos
dados = {
    'Texto': [
        'Promoção imperdível de viagens',
        'Você ganhou um cartão de crédito grátis',
        'Curso de programação em Python',
        'Participação em sorteio de prêmios',
        'Convite para evento de tecnologia',
        'Ganhou bitcoins gratuitos agora'
    ],
    'Spam': [1, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(dados)

# Vetorização do texto
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Texto'])
X_array = X.toarray()
y = df['Spam']

# Treinamento do modelo
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X_array, y)

# Redução de dimensionalidade para visualização
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_array)

# Gráfico com rótulos
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=100)

# Adiciona os rótulos dos e-mails
for i, texto in enumerate(df['Texto']):
    plt.annotate(texto, (X_pca[i, 0], X_pca[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')

# Legenda
handles = [plt.Line2D([], [], marker='o', color='w', label='Não Spam', markerfacecolor='purple', markersize=10),
           plt.Line2D([], [], marker='o', color='w', label='Spam', markerfacecolor='yellow', markersize=10)]
plt.legend(handles=handles)

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Classificação de E-mails com Rótulos')
plt.show()
```
