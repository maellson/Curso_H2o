```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dados simulados com clusters mais prximos
np.random.seed(0)
cluster_1 = np.random.randn(50, 2) + [2, 2]
cluster_2 = np.random.randn(50, 2) + [5, 5]
cluster_3 = np.random.randn(50, 2) + [8, 2]

X = np.vstack((cluster_1, cluster_2, cluster_3))
df = pd.DataFrame(X, columns=['X1', 'X2'])

# Aplicao do K-Means
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['X1', 'X2']])

# Grfico
plt.scatter(df['X1'], df['X2'], c=df['Cluster'], cmap='Accent')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Agrupamento com Dados Mais Prximos')
plt.show()
```
