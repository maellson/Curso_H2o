```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados hipotéticos
dados = {'Area': [50, 60, 70, 80, 90, 100],
         'Preco': [150000, 180000, 210000, 240000, 270000, 300000]}
df = pd.DataFrame(dados)

# Treinamento do modelo
X = df[['Area']]
y = df['Preco']
modelo = LinearRegression()
modelo.fit(X, y)

# Previsões
df['Previsao'] = modelo.predict(X)

# Gráfico
plt.scatter(df['Area'], df['Preco'], color='blue', label='Dados Reais')
plt.plot(df['Area'], df['Previsao'], color='red', label='Linha de Regressão')
plt.xlabel('Área (m²)')
plt.ylabel('Preço (R$)')
plt.title('Regressão Linear - Preço vs. Área')
plt.legend()
plt.show()
```
